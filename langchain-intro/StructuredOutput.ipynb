```python
from dotenv import load_dotenv
load_dotenv()

import os
from langchain.chat_models import init_chat_model

os.environ['GROQ_API_KEY'] = os.getenv("groq_api_key")
model = init_chat_model(model = 'groq:llama-3.1-8b-instant')
```

```python
response = model.invoke('What is the weather in burgess hill')
response.content
```

```python
from pydantic import BaseModel, Field

class Movie(BaseModel):
    title: str = Field(description='The title of the Moview')
    year: int = Field(description='This the release year of a movie')
    director: str = Field(description='Director of the movie')
    rating: float = Field(..., description='Ratings of the movie')

structured_llm = model.with_structured_output(Movie, include_raw=True)
response = structured_llm.invoke('Tell me about Gladiator 2')
response
```

Model with Nested structure response

```python
from pydantic import BaseModel, Field

class Actor(BaseModel):
    name: str
    role: str

class MovieDetail(BaseModel):
    title: str
    year: int
    actors: list[Actor]
    genre: list[str]
    budget: float | None

model_with_structure = model.with_structured_output(MovieDetail)
response = model_with_structure.invoke('Details of inception')
response
```
